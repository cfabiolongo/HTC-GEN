### Identification of the 15 most frequent keywords by area

```python
list_y1 = []
list_y = []
list_keys = []
for index, row in flat_domain.iterrows():    
    list_y.append(row['Y'])
    list_y1.append(row['Y1'])
    keys_total = ''.join(df[(df['Y'] == row['Y']) & (df['Y1'] == row['Y1'])]['keywords'])
    list_keys_total = keys_total.split("; ")
    list_keys_total_min = [s.lower() for s in list_keys_total]
    counts = Counter(list_keys_total_min)
    freq = counts.most_common(15)
    freq = [item[0] for item in freq]    
    stringa = ';'.join(freq)
    list_keys.append(stringa.strip())    

# DataFrame creation
data = {'Y1': list_y1 , 'Y': list_y , 'keywords': list_keys}
df_key = pd.DataFrame(data)
df_key
```

```python
df_key.to_excel("dataset/effective_keys15.xlsx")
```

```python
# Inner join based on Y1, Y
final_df_merged = pd.merge(flat_domain, df_key, on=['Y1', 'Y'], how="inner")
final_df_merged
```

```python
final_df_merged.to_excel("dataset/real_wos_keywords15_data.xlsx")
```

```python
# Split the 'keywords' column and repeat the value of 'area'
split_df = final_df_merged.assign(keywords=final_df_merged['keywords'].str.split(';')).explode('keywords')
split_df
```

```python
conteggi = split_df.groupby(['Y','Y1'])
aree = conteggi.first().reset_index()
aree = aree[['Y','Y1']]

conteggi_raggruppati = conteggi.size().reset_index(name='Conteggio')

# Inner join based on Y, Y1
conteggi_aree = pd.merge(conteggi_raggruppati, aree, on=['Y','Y1'], how="inner")
conteggi_aree
```
