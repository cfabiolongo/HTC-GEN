```python
import pandas as pd
import numpy as np

df = pd.read_excel('dataset/zeroshot_total_keywords_wos.xlsx')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 400)

df
```

```python
with open('results/results_total_keywords_generation.txt', 'r') as file:
    file_content = file.read()

fragments = file_content.split("==================================")
fragments = [fragment.strip() for fragment in fragments if fragment.strip()]
df_file = pd.DataFrame({'Keywords': fragments})
df_file
```

```python
# dataframe reductions
df_redux = df[['Y1', 'Domain', 'Y', 'area']]

merged_df = df_file.merge(df_redux, left_index=True, right_index=True)
merged_df
```

### Single keywords extraction

```python
# Function to extract strings beginning with "1. ", "2. ", etc.

def extract_substrings(keyword_str):
    substrings = []
    keyword_lines = keyword_str.split("\n")
    for line in keyword_lines:
        for i in range(1, 21):
            prefix = f"{i}. "
            if line.startswith(prefix):
                substrings.append(line)
                break
    return substrings

# Above function application to dataframe
merged_df['Extracted_Keywords'] = merged_df['Keywords'].apply(extract_substrings)

# Expand the extracted data into new rows
expanded_data = []
for index, row in merged_df.iterrows():
    for keyword in row['Extracted_Keywords']:
        expanded_data.append({'Y1': row['Y1'], 'Domain': row['Domain'], 'Y': row['Y'], 'area': row['area'], 'Keyword': keyword})

# Create a new DataFrame with the extracted data
new_df = pd.DataFrame(expanded_data)
new_df
```

### Cleaning up Keywords column from spurious elements

```python
new_df['Keyword'] = new_df['Keyword'].str.replace('11. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('12. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('13. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('14. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('15. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('16. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('17. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('18. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('19. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('20. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('1. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('2. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('3. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('4. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('5. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('6. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('7. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('8. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('9. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('10. ', '')
new_df['Keyword'] = new_df['Keyword'].str.replace('"', '')
new_df
```

```python
new_df.to_excel("dataset/lista_total_20keywords.xlsx")
```

```python
conteggi = new_df.groupby(['Y1','Y'])
aree = conteggi.first().reset_index()
aree = aree[['Y1','Domain','Y','area']]

conteggi_raggruppati = conteggi.size().reset_index(name='Conteggio')

# Performing inner join based fields Y1, Y
conteggi_aree = pd.merge(conteggi_raggruppati, aree, on=['Y1','Y'], how="inner")
conteggi_aree
```
