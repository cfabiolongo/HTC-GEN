```python
import pandas as pd
import numpy as np
import re

df = pd.read_excel('dataset/zeroshot_daArea-p1_wos.xlsx')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 400)

df
```

```python
conteggi = df.groupby(['Y', 'Y1'])
aree = conteggi.first().reset_index()
aree = aree[['Y1','Domain','Y','area']]

conteggi_raggruppati = conteggi.size().reset_index(name='Conteggio')

# Executing inner join on Y, Y1
conteggi_aree = pd.merge(conteggi_raggruppati, aree, on=['Y','Y1'], how="inner")
conteggi_aree
```

```python
# Calculating the average value of the "Conteggio" column
media_conteggio = conteggi_raggruppati['Conteggio'].mean()
media_conteggio
```

```python
with open('results/results_abstract_daArea-p1_generation.txt', 'r') as file:
    file_content = file.read()

# Split the content into fragments using the string "================" as a separator
fragments = file_content.split("==================================")

# Remove any empty fragments
fragments = [fragment.strip() for fragment in fragments if fragment.strip()]

# Create a DataFrame with "Abstract" column
df_file = pd.DataFrame({'Abstract': fragments})
df_file

def remove_sure_substrings(text):
    return re.sub(r'Sure.*?\n', '', text)

# Apply the above function to the "Abstract" column
df_file['Abstract'] = df_file['Abstract'].apply(remove_sure_substrings)
df_file
```

```python
# Function to extract and remove the "Title" string
def extract_title(text):
    if "Title:" in text:
        title_start = text.index("Title:") + len("Title:")
        title_end = text.index('\n', title_start)
        title = text[title_start:title_end]
        return title, text.replace("Title:" + title, "").strip()
    else:
        return None, text

# Apply the above function to extract the title and create a new "Title" column
df_file['Title'], df_file['Abstract'] = zip(*df_file['Abstract'].map(extract_title))
df_file
```

```python
# Remove unnecessary white characters from "abstract" column
df_file['Abstract'] = df_file['Abstract'].str.strip()

# Remove the "Abstract" string from the "Abstract" column
df_file['Abstract'] = df_file['Abstract'].str.replace('Abstract:', '')
df_file['Abstract'] = df_file['Abstract'].str.replace('\n', '')

df_file
```

```python
# df reduction
df_redux = df[['Y1', 'Domain', 'Y', 'area']]

# index based merging
merged_df = df_file.merge(df_redux, left_index=True, right_index=True)
merged_df
```

### Extraction of Keywords column from Abstract content

```python
# Estrai il testo dalla colonna "Abstract" e inseriscilo nella colonna "Title"
merged_df['Keywords'] = merged_df['Abstract'].str.extract(r'Keywords: (.*)$')

# Rimuovi le sottostringhe dai titoli nelle colonne "Abstract"
merged_df['Abstract'] = merged_df.apply(lambda row: row['Abstract'].replace("Keywords: " + row['Keywords'], '') if isinstance(row['Keywords'], str) else row['Abstract'], axis=1)
merged_df
```

```python
conteggi = merged_df.groupby(['Y', 'Y1'])
aree = conteggi.first().reset_index()
aree = aree[['Y1','Domain','Y','area']]

conteggi_raggruppati = conteggi.size().reset_index(name='Conteggio')

# inner join based on Y, Y1
conteggi_aree = pd.merge(conteggi_raggruppati, aree, on=['Y','Y1'], how="inner")
conteggi_aree
```

```python
merged_df.to_excel("dataset/zeroshot_daArea_p1_50_finale.xlsx")
```

```python
# Function to select a fraction of the elements for each group
def select_half(group):
    return group.sample(frac=0.1, random_state=42)

# Applicazione della funzione al DataFrame raggruppato
subset_df = merged_df.groupby('Y', group_keys=False).apply(select_half)

# Stampa del risultato
subset_df
```

```python
conteggi = subset_df.groupby(['Y', 'Y1'])
aree = conteggi.first().reset_index()
aree = aree[['Y1','Domain','Y','area']]

conteggi_raggruppati = conteggi.size().reset_index(name='Conteggio')

# inner join based on Y, Y1
conteggi_aree = pd.merge(conteggi_raggruppati, aree, on=['Y','Y1'], how="inner")
conteggi_aree
```

```python
# Calculating the average value of the "Conteggio" column
media_conteggio = conteggi_raggruppati['Conteggio'].mean()
media_conteggio
```

```python
subset_df.to_excel("dataset/zeroshot_daArea_p1_10_finale.xlsx")
```
